## Running Turicreate Models

```python
import turicreate as tc
```

```python
data = tc.SFrame(merge_df)
```

```python
data.show()
```

```python
data.column_names()[1:]
```

```python
data['Location']
```

```python
data['Total'] = data['Home Score'] + data['Away Score']
data.head()
```

```python
train_data, test_data = data.random_split(0.9)
```

```python
model=tc.regression.create(train_data, target='Total', features=['Home Team',
 'Away Team',
 'Total Score Open',                                                               
 'Location',
 'Surface',
 'Roof type',
 'Game Time',
 'Game Day',
 'Month',                                                                         
 'Year'])
```

```python
prediction = model.predict(test_data)
```

```python
prediction
```

```python
model_results = model.evaluate(test_data)
```
